### Load and Clean the dataset


```python
#Import necessary libraries
import pandas as pd

#Load the Titanic dataset
df = pd.read_csv('titanic.csv')

#Display first few rows
print(df.head())
```

```python
#Check basic information and missing values
print('\n Dataset info')
print(df.info())

print('\n Missing vslues in each column:')
print(df.isnull().sum())
```

```python
# Drop columns starting with 'zero'
df = df.loc[:, ~df.columns.str.startswith('zero')]

# Rename the column '2urvived' to 'Survived'
df.rename(columns={'2urvived': 'Survived'}, inplace=True)

# Preview final cleaned dataset
print("\n Cleaned columns:")
print(df.columns.tolist())
print(df.head())

```

###  Descriptive Statistics & Visualizations (Matplotlib & Seaborn)

```python
#Descriptive statistics
print(df.describe(include = 'all'))
```

### Visualizations with Matplotlib & Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Sex_label'] = df['Sex'].map({0: 'Male', 1: 'Female'})

#Distribution of Age
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequancy')
plt.show()
```

```python
#Survival by sex
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue='Survived', data=df, palette='Set2')
plt.title('Survival by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title = 'Survived')
plt.show()
```

```python
#Boxplot of Fare by Pclass
plt.figure(figsize = (7,5))
sns.boxplot(x='Pclass', y ='Fare', data = df, palette='coolwarm')
plt.title('Fare Distribution by Plclass')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize = (10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Between Numerical Features')
plt.show()
```
